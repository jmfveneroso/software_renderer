### Terrain

- The grid is 200 X 200 meters.
- The map has 41 X 41 vertices for a total of 1681 vertices.
- The map has 40 X 40 faces for a total of 1600 squares 
- The tile is 5 by 5 meters.



```python
# Read terrain obj file.

vertices = []
with open('terrain.obj') as f:
    for l in f:
        l = l.strip().split()
        if l[0] == 'v':
            vertices.append([
                round(float(l[1])), 
                round(float(l[3])), 
                round(float(l[2]) - 3.295138, 1)
            ])

print(len(vertices), 'vertices.')

grid = np.zeros((41, 41))
vertices.sort(key=lambda v: v[0])
vertices.sort(key=lambda v: v[1])

i = 0
j = 0
for v in vertices:
    grid[i][j] = v[2]
    j += 1
    if j == 41:
        j = 0
        i += 1
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

min_val, max_val = 0, 15

intersection_matrix = np.random.randint(0, 10, size=(max_val, max_val))

ax.matshow(grid, cmap=plt.cm.Blues)
```

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

%matplotlib inline

x = np.arange(0,41,1)
y = np.arange(0,41,1)
X,Y = np.meshgrid(x,y)
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')
ax.set_zlim(0, 80)
ax.plot_surface(X, Y, grid)
plt.show()
```

### Write binary file

```python
import struct

with open('terrain.data', 'w') as f:
    f.write(str(41)+'\n')
    for i in range(41):
        for j in range(41):
            f.write(str(grid[i][j])+'\n')
```
